```python
import matplotlib.pyplot as plt
import re

def plot_losses_from_file(file_path):
    # Initialize lists to store losses and epochs
    epochs = []
    avg_first_20_losses = []
    avg_last_20_losses = []

    # Regular expression to extract loss values
    loss_pattern = re.compile(r"Epoch (\d+): Avg First 20 Loss: ([\d.]+), Avg Last 20 Loss: ([\d.]+)")

    # Open and parse the file
    with open(file_path, 'r') as file:
        for line in file:
            match = loss_pattern.match(line)
            if match:
                epoch = int(match.group(1))
                first_20_loss = float(match.group(2))
                last_20_loss = float(match.group(3))
                
                epochs.append(epoch)
                avg_first_20_losses.append(first_20_loss)
                avg_last_20_losses.append(last_20_loss)

    # Plot the losses
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, avg_first_20_losses, label="Avg First 20 Loss")
    plt.plot(epochs, avg_last_20_losses, label="Avg Last 20 Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Losses vs. Epochs")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
file_path = "../outputs/_run_2/training_log.txt"
plot_losses_from_file(file_path)
```
